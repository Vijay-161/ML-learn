```python
import pandas as pd
df = pd.read_csv('one_hot_encoding.csv')
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
final = merged.drop(['town','Janakpur'], axis= 'columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = final.drop('price', axis='columns')
x
```

```python
y = final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,1,0]])
```

```python
model.predict([[2600,1,0]])
```

```python
model.score(x,y)
```
